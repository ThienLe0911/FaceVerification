# Face Verification - Fine-tuning on Google Colab

**⚠️ This notebook is designed to run on Google Colab with GPU acceleration.**

This notebook covers:

1. **Environment Setup** - Installing requirements and mounting Google Drive
2. **Dataset Preparation** - Loading and organizing training data
3. **Model Configuration** - Setting up FaceNet for fine-tuning
4. **Training Pipeline** - Light fine-tuning with custom data

## Instructions for Google Colab

1. **Upload this notebook** to Google Colab
2. **Enable GPU** in Runtime → Change runtime type → GPU
3. **Mount Google Drive** to access your dataset
4. **Follow the cells** in order, paying attention to file paths

## Prerequisites

- Google Colab account with GPU access
- Dataset uploaded to Google Drive
- Basic understanding of PyTorch training loops
